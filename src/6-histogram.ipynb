### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Data for plots
x = np.linspace(0, 10, 100)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Line chart
axes[0, 0].plot(x, y, color='b', linewidth=2)
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X Values")
axes[0, 0].set_ylabel("Y Values")
axes[0, 0].grid(True)

# Bar chart
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot
axes[1, 0].scatter(scatter_x, scatter_y, color='blue')
for i in range(len(scatter_x)):
    axes[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X Values")
axes[1, 0].set_ylabel("Y Values")
axes[1, 0].grid(True)

# Histogram
axes[1, 1].hist(data, bins=5, color='cyan', edgecolor='black', alpha=0.7)
axes[1, 1].set_title("Histogram")
axes[1, 1].set_xlabel("Data Values")
axes[1, 1].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
